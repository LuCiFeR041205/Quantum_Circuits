```python
from qiskit.quantum_info import SparsePauliOp , Statevector
from qiskit_aer import Aer
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
import matplotlib as plt 
```

```python
qc = QuantumCircuit(2)
pauli_op = SparsePauliOp.from_list([("XZ" ,1.0)])
qc.x(0)
qc.z(1)
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc)
result = job.result()
statevector = result.get_statevector()
print(statevector)
plot_bloch_multivector(qc)
```

```python

expectation = statevector.expectation_value(pauli_op)
print("Expectation value of XZ:", expectation)
```

```python
qc1 = QuantumCircuit(2)
pauli_op1 = SparsePauliOp.from_list([("ZZ" , 1.0),("XI" , 1.0)])
qc1.h(0)
qc1.cx(0,1)
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc1)
result = job.result()
statevector1 = result.get_statevector()
print(statevector1)
expectation1 = statevector1.expectation_value(pauli_op1)
print("expectation value of ZZ+IX:", expectation1)
```
