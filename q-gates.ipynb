```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
circuit = QuantumCircuit(1,1)
circuit.h(0)
circuit.measure(0,0)
(circuit.draw('mpl'))
```

```python
circuit = QuantumCircuit(1,1)
circuit.x(0)
circuit.measure(0,0)
(circuit.draw('mpl'))
```

```python
circuit = QuantumCircuit(1,1)
circuit.z(0)
circuit.measure(0,0)
(circuit.draw('mpl'))
```

```python
circuit = QuantumCircuit(2,2)
circuit.x(1)
circuit.cx(0,1)
circuit.measure([0,1] , [0,1])
(circuit.draw('mpl'))
```

```python
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
circuit.measure([0,1] , [0,1])
(circuit.draw('mpl'))
```
